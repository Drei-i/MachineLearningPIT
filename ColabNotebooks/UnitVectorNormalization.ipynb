```python
import pandas as pd
from sklearn.preprocessing import Normalizer

# Load Dataset
from google.colab import files

# Upload Dataset Manually
uploaded = files.upload()

# Read Uploaded CSV File
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

numerical_columns = [
    'work_hours',
    'screen_time_hours',
    'meetings_count',
    'breaks_taken',
    'after_hours_work',
    'sleep_hours',
    'task_completion_rate',
    'burnout_score'
]

# Apply Unit Vector Normalization
normalizer = Normalizer(norm='l2')

df_unit = df.copy()
df_unit[numerical_columns] = normalizer.fit_transform(df[numerical_columns])

print(df_unit.head())
```
